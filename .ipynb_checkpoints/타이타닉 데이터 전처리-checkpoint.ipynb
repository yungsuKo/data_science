```python
%matplotlib inline
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_csv("./titanic/train.csv")
test_df = pd.read_csv("./titanic/test.csv")
```

```python
all_df = pd.concat([train_df, test_df], sort=False).reset_index(drop=True)
```

```python
all_df.isnull().sum()
```

```python
# 결측치 메우기
    # Fare의 결측치를 메우기 위해서 pClass 별 fare의 평균을 구한다.
fare_mean = all_df[["Pclass", "Fare"]].groupby("Pclass").mean().reset_index()
fare_mean.columns = ["Pclass", "Fare_mean"]
fare_mean
```

```python
all_df = pd.merge(all_df, fare_mean, on="Pclass", how="left")
all_df.loc[(all_df['Fare'].isnull()), "Fare"] = all_df["Fare_mean"]
all_df = all_df.drop(["Fare_mean_x", "Fare_mean_y"], axis=1)
all_df
```

```python
all_df.isnull().sum()
```

```python
### 호칭을 변수로 추가하기
all_df["Name"].head()
```

```python
# name_df=all_df["Name"].str.replace(",", ".").str.split(".", 2)
name_df=all_df["Name"].str.split('[,.]', expand=True)
```

```python
name_df.columns = ["family_name", "honorific", "name", "_"]
```

```python
name_df["family_name"] = name_df["family_name"].str.strip()
name_df["honorific"] = name_df["honorific"].str.strip()
name_df["name"] = name_df["name"].str.strip()
```

```python
name_df["honorific"].value_counts()
```

```python
all_df = pd.concat([all_df, name_df], axis=1)
```

```python
all_df
```

```python
plt.figure(figsize=(18, 5))
sns.boxplot(x="honorific", y="Age", data=all_df)
```

```python
all_df[["honorific", "Age"]].groupby("honorific").mean()
```

```python
train_df = pd.concat([train_df, name_df[0:len(train_df)].reset_index(drop=True)], axis=1)
test_df = pd.concat([test_df, name_df[len(train_df):].reset_index(drop=True)], axis=1)
```

```python
honorific_df = train_df[["honorific", "Survived", "PassengerId"]].dropna().groupby(["honorific", "Survived"]).count().unstack()
honorific_df.plot.bar(stacked=True)
```

```python
honorific_age_min = all_df[["honorific", "Age"]].groupby("honorific").mean()
all_df = pd.merge(all_df , honorific_age_min, on="honorific", how="left")
all_df
```

```python
all_df.loc[(all_df["Age_x"].isnull()),"Age_x"]=all_df["Age_y"]
all_df = all_df.drop("Age_y", axis = 1)
all_df
```

```python
all_df = all_df.drop("_", axis = 1)
```

```python
categories = all_df.columns[all_df.dtypes == "object"]
print(categories)
```

```python
all_df.loc[~((all_df["honorific"] == "Mr") | (all_df["honorific"] == "Miss") | (all_df["honorific"] == "Mrs")|(all_df["honorific"] == "Master")),"honorific"] = "others"
all_df



```
