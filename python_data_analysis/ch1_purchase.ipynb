```python
import pandas as pd
customer_master = pd.read_csv('ch1_data/customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('ch1_data/item_master.csv')
transaction_1 = pd.read_csv('ch1_data/transaction_1.csv')
transaction_detail_1 = pd.read_csv('ch1_data/transaction_detail_1.csv')
transaction_2 = pd.read_csv('ch1_data/transaction_2.csv')
transaction_detail_2 = pd.read_csv('ch1_data/transaction_detail_2.csv')
```

```python
transaction_detail_2.head()
```

```python
# 데이터 결합
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
# 결합 여부 검증
print(len(transaction_1))
print(len(transaction_2))
print(len(transaction))
```

```python
join_data = pd.merge(
    transaction_detail_1, 
    transaction[["transaction_id", "payment_date", "customer_id"]],
    on="transaction_id", 
    how="left"
)
join_data.head()
```

```python
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```
