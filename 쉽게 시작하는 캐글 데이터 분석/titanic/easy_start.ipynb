```python
train_df_corr = pd.get_dummies(train_df, columns=["Embarked"])
```

```python
train_df_corr = train_df_corr.drop(['Ticket', 'Cabin'], axis=1)
```

```python
train_df_corr_name = train_df_corr["Name"].str.split("[,.]",n=2, expand=True)
```

```python
train_df_corr["honority"] = train_df_corr_name[1]
```

```python
train_df_corr.head()
```

```python
train_df_corr.isnull().sum()
```

```python
train_df_corr_avg = train_df_corr[["honority", "Age"]].groupby("honority").sum()/train_df_corr[["honority", "Age"]].groupby("honority").count()
```

```python
train_df_corr_avg
```

```python
train_df_corr =pd.merge(train_df_corr, train_df_corr_avg, on="honority")
```
